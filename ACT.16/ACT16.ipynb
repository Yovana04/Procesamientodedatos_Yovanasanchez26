```python
import random

class Individuo:
    def __init__(self, cantidad_genes=5):
        self.genes = [random.randint(1, 10) for _ in range(cantidad_genes)]
        self.fitness = 0

    def evaluar(self):
        self.fitness = sum(self.genes)

    def mutar_gen(self):
        posicion = random.randint(0, len(self.genes) - 1)
        self.genes[posicion] = random.randint(1, 10)

    def __str__(self):
        return f"Individuo(genes={self.genes}, fitness={self.fitness})"


class Poblacion:
    def __init__(self, total_individuos=5, cantidad_genes=5):
        self.individuos = [Individuo(cantidad_genes) for _ in range(total_individuos)]

    def evaluar_individuos(self):
        for individuo in self.individuos:
            individuo.evaluar()

    def obtener_mejor(self):
        return max(self.individuos, key=lambda ind: ind.fitness)

    def imprimir(self):
        for indice, individuo in enumerate(self.individuos, 1):
            print(f"{indice}. {individuo}")

poblacion_prueba = Poblacion(total_individuos=6, cantidad_genes=4)
poblacion_prueba.evaluar_individuos()

print("Listado de individuos:")
poblacion_prueba.imprimir()

mejor = poblacion_prueba.obtener_mejor()
print("\nMejor individuo encontrado:")
print(mejor)

print("\nAplicando al mejor individuo...")
mejor.mutar_gen()
mejor.evaluar()

print("Resultado después de la mutación:")
print(mejor)
```
