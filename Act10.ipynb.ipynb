```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns

df_temperatura = pd.read_csv("celsius_a_fahrenheit.csv")
df_temperatura.head()
```

```python
print(df_temperatura.head())
```

```python
sns.scatterplot(x='Celsius', y='Fahrenheit', data=df_temperatura)
plt.title("Relación entre Celsius y Fahrenheit")
plt.show()
```

```python
X_train = df_temperatura['Celsius'].values
y_train = df_temperatura['Fahrenheit'].values
```

```python
modelo_lineal = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)   # activación lineal por defecto
])

modelo_lineal.compile(
    optimizer=tf.keras.optimizers.Adam(1.0),
    loss='mean_squared_error'
)

hist_lineal = modelo_lineal.fit(X_train, y_train, epochs=200, verbose=False)
```

```python
plt.plot(hist_lineal.history['loss'])
plt.title("Pérdida del modelo con activación lineal")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
temp_c = np.array([100.0])
pred_lineal = modelo_lineal.predict(temp_c)

print("Predicción con activación lineal:", pred_lineal[0][0])
print("Resultado real:", (100 * 9/5) + 32)
```

```python
modelo_relu = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1, activation='relu')
])

modelo_relu.compile(
    optimizer=tf.keras.optimizers.Adam(1.0),
    loss='mean_squared_error'
)

hist_relu = modelo_relu.fit(X_train, y_train, epochs=200, verbose=False)
```

```python
plt.plot(hist_relu.history['loss'])
plt.title("Pérdida del modelo con activación ReLU")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
pred_relu = modelo_relu.predict(temp_c)

print("Predicción con activación ReLU:", pred_relu[0][0])
print("Resultado real:", (100 * 9/5) + 32)
```

```python
print("Comparación de resultados")
print("Lineal:", pred_lineal[0][0])
print("ReLU:", pred_relu[0][0])
print("Valor real:", (100 * 9/5) + 32)
```

```python
modelo_mae = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

modelo_mae.compile(
    optimizer=tf.keras.optimizers.Adam(1.0),
    loss='mean_absolute_error'
)

hist_mae = modelo_mae.fit(X_train, y_train, epochs=200, verbose=False)

plt.plot(hist_mae.history['loss'])
plt.title("Pérdida usando MAE")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
