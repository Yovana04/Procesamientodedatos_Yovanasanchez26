## Ejercicio 2: Éxito de una misión

```python

import pandas as pd
import numpy as np
pd.set_option("display.width", 140)
```

```python

vals=["Bajo","Medio","Alto"]
rng=np.random.default_rng(7)

def exito(n,s):
    return "Sí" if rng.random() < (0.3 if n=="Bajo" else 0.6 if n=="Medio" else 0.85) else "No"

rows=[]
for _ in range(500):
    NE=rng.choice(vals); EQ=rng.choice(vals); H=rng.choice(vals)
    S=rng.choice(vals); EM=rng.choice(["Positivo","Neutral","Negativo"])
    rows.append([NE,EQ,H,S,EM,exito(NE,S)])

df=pd.DataFrame(rows,columns=["NivelExp","Equipamiento","Habilidades","Salud","EstadoMental","Exito"])
df.head()
```

```python

for col in ["NivelExp","Equipamiento","Habilidades","Salud","EstadoMental"]:
    print(f"\nP(Éxito | {col})")
    display(pd.crosstab(df[col],df["Exito"],normalize="index").round(3))
```
