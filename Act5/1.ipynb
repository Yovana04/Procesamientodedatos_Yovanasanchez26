## Ejercicio 1: Compra de un videojuego

```python

import pandas as pd
import numpy as np
pd.set_option("display.width", 140)
```

```python

vals_P = ["Alto","Medio","Bajo"]
vals_G = ["Acción","Aventura","Estrategia","Otros"]
vals_CP = ["Alta","Baja"]
vals_Pl = ["Consola","PC"]
vals_A = ["Sí","No"]

rng = np.random.default_rng(42)

def compra(P, CP, A):
    score = {"Alto":-1,"Medio":0,"Bajo":1}[P] + {"Alta":1,"Baja":-1}[CP] + {"Sí":1,"No":0}[A]
    return "Sí" if rng.random() < 1/(1+np.exp(-score)) else "No"

rows=[]
for _ in range(500):
    P=rng.choice(vals_P); G=rng.choice(vals_G)
    CP=rng.choice(vals_CP); Pl=rng.choice(vals_Pl); A=rng.choice(vals_A)
    rows.append([P,G,CP,Pl,A,compra(P,CP,A)])

df=pd.DataFrame(rows,columns=["Precio","Genero","Criticas","Plataforma","Amigos","Compra"])
df.head()
```

```python

for col in ["Precio","Genero","Criticas","Plataforma","Amigos"]:
    print(f"\nP(Compra | {col})")
    display(pd.crosstab(df[col],df["Compra"],normalize="index").round(3))
```
