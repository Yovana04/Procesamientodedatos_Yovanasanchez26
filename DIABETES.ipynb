```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = load_diabetes()
X = data.data[:, np.newaxis, 0]   
y = data.target                  
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
print("MSE test:", round(mean_squared_error(y_test, y_pred_test), 2))
print("R2 test:", round(r2_score(y_test, y_pred_test), 2))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

ax[0].scatter(X_train, y_train)
ax[0].plot(X_train, y_pred_train)
ax[0].set_title("Training set")
ax[0].set_xlabel("Feature (as carat)")
ax[0].set_ylabel("Target (as price)")

ax[1].scatter(X_test, y_test)
ax[1].plot(X_train, y_pred_train)
ax[1].set_title("Testing set")
ax[1].set_xlabel("Feature (as carat)")

plt.suptitle("Linear Regression Model")
plt.show()
```
