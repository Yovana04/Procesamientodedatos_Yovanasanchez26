```python
import numpy as np
import matplotlib.pyplot as plt

ingresos = np.array([1452,1361,1426,1470,1456,1430,1354,1442,1394,1459,
                     1399,1458,1537,1425,1445,1439,1348,1450,1431,1446,
                     1485,1405,1461,1490,1426])

habitaciones = np.array([23,47,21,39,37,29,23,44,45,16,
                         30,42,54,27,34,15,19,38,44,47,
                         43,38,51,61,39])

correlacion = np.corrcoef(habitaciones, ingresos)[0,1]

print("Coeficiente de correlación:", correlacion)

# Graficar
plt.scatter(habitaciones, ingresos)
plt.xlabel("Habitaciones ocupadas")
plt.ylabel("Ingreso del restaurante")
plt.title("Gráfica de correlación lineal simple")
plt.show()
```

```python
m, b = np.polyfit(habitaciones, ingresos, 1)

plt.scatter(habitaciones, ingresos)
plt.plot(habitaciones, m*habitaciones + b)
plt.xlabel("Habitaciones ocupadas")
plt.ylabel("Ingreso del restaurante")
plt.title("Correlación con línea de regresión")
plt.show()
```

```python
print("Existe una correlación positiva moderada entre habitaciones ocupadas e ingreso.")
```
